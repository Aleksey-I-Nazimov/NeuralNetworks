```python
from torch import optim
from torch import nn
import torch
import torch.nn.functional as F
import torchvision

import matplotlib.pyplot as plt
import numpy as np
import math
import gc

import torchvision.transforms as transforms
from torchvision.transforms.functional import to_grayscale
```

```python
# https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html



#  in_channels (int) – Number of channels in the input image

#  out_channels (int) – Number of channels produced by the convolution

#  kernel_size (int or tuple) – Size of the convolving kernel

#  stride (int or tuple, optional) – Stride of the convolution. Default: 1

#  padding (int, tuple or str, optional) – Padding added to all four sides of the input. Default: 0

#  padding_mode (str, optional) – 'zeros', 'reflect', 'replicate' or 'circular'. Default: 'zeros'

#  dilation (int or tuple, optional) – Spacing between kernel elements. Default: 1

#  groups (int, optional) – Number of blocked connections from input channels to output channels. Default: 1

#  bias (bool, optional) – If True, adds a learnable bias to the output. Default: True

#  groups controls the connections between inputs and outputs.
#  in_channels and out_channels must both be divisible by groups. For example,
#  At groups=1, all inputs are convolved to all outputs.
#  At groups=2, the operation becomes equivalent to having two conv layers side by side,
#  each seeing half the input channels and producing half the output channels, and both subsequently concatenated.
#  At groups= in_channels, each input channel is convolved with its own set of filters
#  (of size out_channelsin_channelsin_channelsout_channels​).

conv_layer = nn.Conv2d(in_channels=3,
                       out_channels=3,
                       kernel_size=(3,3),
                       stride=2,
                       padding=(1,1),
                       padding_mode='replicate',
                       bias=True,
                       groups=3,
                       dilation=1)
x=torch.rand(48).reshape((3,4,4))
print (x)

y = conv_layer(x)
print (y)
```

```python

def execute (runnable):
    if torch.cuda.device_count()>0:
        print ("  --> Let's work with cuda")
        device = torch.device("cuda:0")
        gc.collect()
        torch.cuda.empty_cache()
        if runnable is not None:
            with torch.cuda.device(0):
                return runnable(device)
        return device
    else:
        print ("  --> Let's work with CPU")
        device = torch.device("cpu")
        if runnable is not None:
            return runnable(device)
        return device
```

```python
IMG_SIZE = 32
COLOR_CHANNELS = 1

class DatasetAdapter(torch.utils.data.Dataset):

    def __init__(self, dataset, from_cnt=0, to_cnt=1000, channel_num=1,size=32, is_gray_scale=True):
        self.dataset = dataset
        self.from_cnt = from_cnt
        self.to_cnt = to_cnt
        self.channel_num = channel_num
        self.size=size
        self.is_gray_scale = is_gray_scale

    def __len__(self):
        # TODO: Let's assert the negative values
        return self.to_cnt - self.from_cnt

    def __getitem__(self, idx):
        biased_index = self.from_cnt+idx
        image, label = self.dataset[biased_index]
        
        if self.is_gray_scale == True:
            image = to_grayscale(img=image, num_output_channels=self.channel_num)

        image = transforms.ToTensor()(image)
        image = image/image.max()
        image = image.reshape(self.channel_num, self.size, self.size)
        return image,label
```

```python
# https://pytorch.org/vision/stable/generated/torchvision.datasets.CIFAR10.html
# https://www.cs.toronto.edu/~kriz/cifar.html

train_dataset = torchvision.datasets.CIFAR10(root='cifar10/', train=True, download=True)
test_dataset = torchvision.datasets.CIFAR10(root='cifar10_test/', train=False, download=True)

TRAIN_SIZE = 20000
train_set = DatasetAdapter(dataset=train_dataset,
                           from_cnt=0,to_cnt=TRAIN_SIZE,channel_num=1,
                          size=IMG_SIZE,is_gray_scale=True)

TEST_SIZE = 1000
test_set = DatasetAdapter(dataset=test_dataset,
                           from_cnt=0,to_cnt=TEST_SIZE,channel_num=1,
                          size=IMG_SIZE,is_gray_scale=True)
```

```python
test_set_loader = torch.utils.data.DataLoader(test_set,
                                              batch_size=TEST_SIZE,
                                              shuffle=False,
                                              num_workers=0)

CLASSES = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

number=50
for batch, label in test_set_loader:
    print(batch.shape)
    plt.imshow(batch[number].view(COLOR_CHANNELS, IMG_SIZE, IMG_SIZE).permute(1, 2, 0))
    print('label is', CLASSES[label[number]])
    break
```

```python

class ConvolutionPlusPerceptron(nn.Module):
    
    def __init__(self, input_size, channel_num, kernel_size, conv1_nn_size, conv_layer_num, hidden_nn_size, hidden_layer_num, output_size):
        
        super().__init__()
        self.layers = []

        # Making the convolution layers:-------------------------------
        dropout_base = 0.05
        for cnt in range(0,conv_layer_num):
            if cnt==0:
                conv_input_size=channel_num
            else:
                conv_input_size=conv1_nn_size
                
            self.layers.append(make_conv2d(in_channels=conv_input_size,layer_size=conv1_nn_size,
                                           stride=1,kernel_size=kernel_size,groups=channel_num))
            self.layers.append(nn.Dropout( dropout_base ))
            self.layers.append(nn.BatchNorm2d(conv1_nn_size))
            
            self.layers.append(make_conv2d(in_channels=conv1_nn_size,layer_size=conv1_nn_size,
                                           stride=2,kernel_size=kernel_size,groups=channel_num))
            self.layers.append(nn.Dropout( dropout_base ))
            self.layers.append(nn.BatchNorm2d(conv1_nn_size))
        
        # Making the basic perceptron:---------------------------------
        scale_coef = math.pow(2,conv_layer_num)
        #conv_nodes = int(conv1_nn_size*scale_coef)
        conv_size = int(input_size/scale_coef)
        perceptron_input_size = conv1_nn_size*conv_size*conv_size*channel_num
        
        self.layers.append(nn.Linear(perceptron_input_size,hidden_nn_size))
        self.layers.append(nn.Dropout( dropout_base ))
        
        for cnt in range(0,hidden_layer_num):
            self.layers.append(nn.Linear(hidden_nn_size,hidden_nn_size))
            self.layers.append(nn.BatchNorm1d(hidden_nn_size))
            drop_coef = (1.0/hidden_layer_num) * cnt + dropout_base
            drop_coef = 0.50 if drop_coef>0.50 else drop_coef
            self.layers.append(nn.Dropout( drop_coef ))
        self.layers.append(nn.Linear(hidden_nn_size,output_size))

        # Registering modules for torch:-------------------------------
        for cnt,layer in enumerate(self.layers):
            key = "layer{}".format(cnt)
            self.add_module(key,layer)

    def forward(self, xtr):
        prev_layer=None
        cur_layer=None
        for cnt,layer in enumerate(self.layers):
            if isinstance(layer,nn.Linear)==True or isinstance(layer,nn.Conv2d)==True:
                prev_layer=cur_layer
                cur_layer=layer
                if isinstance(prev_layer,nn.Conv2d) and isinstance(cur_layer,nn.Linear):
                    xtr=xtr.reshape((xtr.shape[0],xtr.shape[1]*xtr.shape[2]*xtr.shape[3]))
                if cnt<len(self.layers)-1:
                    xtr = F.leaky_relu(layer(xtr))
                else :
                    xtr = F.sigmoid(layer(xtr))
            else:
                xtr = layer(xtr)
        return xtr
    
    def predict_by_softmax(self, x):
        self.eval()
        xtr,values,indexes=None,None,None
        with torch.no_grad():
            xtr = self.forward(x)
            xtr = nn.Softmax(dim=1)(xtr)
        values,indexes = torch.max(xtr, 1)
        self.train()
        return indexes
```

```python
network = ConvolutionPlusPerceptron (input_size=IMG_SIZE,
                                     channel_num=COLOR_CHANNELS,
                                     kernel_size=3,
                                     conv1_nn_size=24,
                                     conv_layer_num=4,
                                     hidden_nn_size=128,
                                     hidden_layer_num=4,
                                     output_size=len(CLASSES)).to(execute(None))

for idx, module in enumerate(network.modules()):
     print(idx, '->', module)

criteria = nn.MSELoss()
optimizer = optim.SGD(network.parameters(), lr=0.005, momentum=0.002)
```

```python
def to_device(data, device):
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

def load_to_device(loader,device):
    load_list=[]
    for i, data in enumerate(loader):
        load_list.append((to_device(data[0],device), to_device(data[1],device)))
    return load_list

BATCH_SIZE = int(TRAIN_SIZE/10)

train_set_loader = torch.utils.data.DataLoader(
        train_set,batch_size=BATCH_SIZE,shuffle=True)
test_set_loader = torch.utils.data.DataLoader(
        test_set, batch_size=TEST_SIZE,shuffle=True)

data_dev = execute(None)
train_list = load_to_device(train_set_loader,data_dev)
test_list = load_to_device(test_set_loader,data_dev)
```

```python
# https://discuss.pytorch.org/t/error-while-multiprocessing-in-dataloader/46845/11
# from tqdm.notebook import trange, tqdm

def adapt_perceptron (device):
    epoch_max_num = 125000
    history=[]
    max_test_tp=0
    for epoch_num in range(epoch_max_num):
        running_loss = 0.0
        for train_batch in train_list:
            x_train, y_train = train_batch[0], train_batch[1]
            y_train = F.one_hot(num_classes=10,input=y_train).float();
            y_train[y_train==1.0]=0.75
            optimizer.zero_grad()
            y_pred = network(x_train)
            loss = criteria(y_pred, y_train)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        test_tp=0
        test_overall=0
        y_stat = torch.zeros ((len(test_list),TEST_SIZE),device=device).long()
        y_stat[y_stat==0]=-1
        for j,data in enumerate(test_list):
            x_test, y_test = data[0], data[1]
            y_pred = network.predict_by_softmax(x_test)
            test_overall += len(y_test)
            test_tp += (y_test == y_pred).sum()
            xtr = y_test!=y_pred
            y_stat[j,xtr]=y_test[xtr]

        history.append((epoch_num,test_tp,running_loss))
        if epoch_num % 100==0 or max_test_tp<test_tp:
            if max_test_tp<test_tp:
                max_test_tp=test_tp
                torch.save(network.state_dict(), "model-{}.torch".format(test_tp))
                print ("  >> Saved: {}".format(test_tp))
                
            print("Epoch [{}/{}] loss: {} TP={} Overall={} max={}".format(epoch_num,epoch_max_num,
                                                                          running_loss,test_tp,
                                                                          test_overall,max_test_tp))
            (class_nums,failures)=torch.unique(input=y_stat[y_stat>=0],sorted=True,return_counts=True)
            print("  >> NR: ",class_nums,failures)
            
    print('Adapting is finished!')
    return history

train_history = execute(runnable=adapt_perceptron)
```

```python
epoch_list=[]
tp_list=[]
mse_list=[]
for (num,tp,mse) in train_history:
    epoch_list.append(num)
    tp_list.append(tp.to(device='cpu').item())
    mse_list.append(mse)

plt.plot(epoch_list,tp_list)
plt.show()
plt.plot(epoch_list,mse_list)
    
```

```python

# Loading TEST-LIST:-----------------------------------------------------------------------
test_dataset = torchvision.datasets.CIFAR10(root='cifar10_test/',
                                            train=False, download=True)

test_adapter = DatasetAdapter(dataset=test_dataset,
                          from_cnt=0,to_cnt=10000,channel_num=1,
                          size=IMG_SIZE,is_gray_scale=True)

full_test_set_loader = torch.utils.data.DataLoader(
        test_adapter, batch_size=1000,shuffle=True)

cpu_dev = torch.device("cpu")
test_list = load_to_device(full_test_set_loader,cpu_dev)
#--------------------------(loading test list)----------------------------------------------


# Reading the model:------------------------------------------------------------------------
test_model = ConvolutionPlusPerceptron (input_size=IMG_SIZE,
                                     channel_num=COLOR_CHANNELS,
                                     kernel_size=3,
                                     conv1_nn_size=24,
                                     conv_layer_num=4,
                                     hidden_nn_size=128,
                                     hidden_layer_num=4,
                                     output_size=len(CLASSES)).to(cpu_dev)

test_model.load_state_dict(torch.load("model-685.torch"))
test_model.to(cpu_dev)
test_model.eval()
#--------------------------(reading the model)-----------------------------------------------


# Testing:-----------------------------------------------------------------------------------
def make_test (device):
    test_tp=0
    test_overall=0
    y_stat = torch.zeros ((len(test_list),TEST_SIZE),device=device).long()
    y_stat[y_stat==0]=-1
    
    for j,data in enumerate(test_list):
        x_test, y_test = data[0], data[1]
        y_pred = test_model.predict_by_softmax(x_test)
        test_overall += len(y_test)
        test_tp += (y_test == y_pred).sum()
        xtr = y_test!=y_pred
        y_stat[j,xtr]=y_test[xtr]

    (class_nums,failures)=torch.unique(input=y_stat[y_stat>=0],sorted=True,return_counts=True)

    return (test_tp,test_overall,class_nums,failures)

result = make_test(cpu_dev)
print (result)
```
