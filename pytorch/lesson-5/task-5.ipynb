```python
import os

root_path = "../../../DataSetStore/set_lipstick_original"
#root_path = "/home/alex/dev/AiLearning/DataSetStore/set_lipstick_original"

class Meta:
    def __init__(self, image_file, mask_file, width, height,x_min,x_max,y_min,y_max):
        self.image_path = self.update_image_path(file_name=image_file)
        self.mask_path = self.update_mask_path(file_name=mask_file)
        self.width = int(width)
        self.height = int(height)
        self.x_mask_min = int(x_min)
        self.x_mask_max = int(x_max)
        self.y_mask_min = int(y_min)
        self.y_mask_max = int(y_max)

    def update_image_path (self,file_name):
        return os.path.join(root_path,"720p",file_name)

    def update_mask_path (self,file_name):
        return os.path.join(root_path,"mask",file_name)

    def calculate_suitable_size(self,lips_per_face):
        delta_x = self.x_mask_max-self.x_mask_min
        delta_y = self.y_mask_max-self.y_mask_min

        x1 = self.check_value(self.x_mask_min-delta_x)
        y1 = self.check_value(self.y_mask_min-int(lips_per_face*delta_y))
        x2 = self.check_value(self.x_mask_max+delta_x,self.width)
        y2 = self.check_value(self.y_mask_max+delta_y,self.height)
        return (y1,y2,x1,x2)

    def check_value(self,value,limit=None):
        if limit==None:
            if value<0:
                return 0
            else: return value
        else:
            if value>limit:
                return limit-1
            else:
                return value

    def __str__(self):
        max_str="coordinates: {} {} {} {}".format(self.x_mask_min,self.x_mask_max,self.y_mask_min,self.y_mask_max)
        return "Meta[image_path={},mask_path={},width={},height={},{}]".format(self.image_path,
                                                                            self.mask_path,
                                                                            self.width,
                                                                            self.height,
                                                                            max_str)


csv_path = os.path.join(root_path,"list.csv")
meta_list = []
with open(csv_path) as csv_file:
    file_lines = csv_file.read().splitlines()
    for cnt,line in enumerate(file_lines):
        if cnt==0:
            #print (line)
            continue
        #print (line)
        segments = line.split(",")
        meta=Meta(image_file=segments[0],
                  mask_file=segments[8],
                  width=segments[1],
                  height=segments[2],
                  x_min=segments[4],
                  x_max=segments[6],
                  y_min=segments[5],
                  y_max=segments[7])
        meta_list.append(meta)
        #print(meta)
        #break
```

```python
        
import cv2
import matplotlib.pyplot as plt
from pathlib import Path


def load_image_by_path(file_path,verbose=True):
    path = Path(file_path)
    if verbose :
        print ("  --> Loading {}".format(path.stat()))
        
    image= cv2.imread(str(path), cv2.IMREAD_UNCHANGED)
    
    if verbose :
        print (image.shape)
    return image

def to_gray_and_crop(image,rect,verbose=True):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    if verbose:
        print ("  --> Cropping: {}".format(rect))
        
    gray= gray[rect[0]:rect[1],rect[2]:rect[3]]
    return gray

def decrese_rect_size(image,rect_size,verbose=True):
    new_size = (rect_size,rect_size)
    if verbose:
        print ("  --> Downscaling: from {} to {}".format(image.shape,new_size))
    new_image = cv2.resize(image, new_size, interpolation = cv2.INTER_LINEAR)
    return new_image

def process_image (image_path,best_rect,best_size,verbose):
    image = load_image_by_path(image_path,verbose)
    image = to_gray_and_crop(image,best_rect,verbose)
    image = decrese_rect_size(image,best_size,verbose)
    return image

def plot_ioo (input_img,output_img_1,output_img_2):
    
    if output_img_2 is not None:
        fig,ax = plt.subplots(1,3)
    else:
        fig,ax = plt.subplots(1,2)

    ax[0].imshow(input_img)
    ax[0].set_title('input_img')
    ax[1].imshow(output_img_1)
    ax[1].set_title('output_img_1')
    
    if output_img_2 is not None:
        ax[2].imshow(output_img_2)
        ax[2].set_title('output_img_2')
    plt.show()


LIPS_PER_FACE = 4.5
IMG_SIZE = 64
CHANNEL_NUM = 1
VERBOSE = False

X = []
Y = []

for cnt,meta in enumerate(meta_list):
    if VERBOSE:
        print (meta)
    rect = meta.calculate_suitable_size(lips_per_face=LIPS_PER_FACE)
    try:
        input_image = process_image(meta.image_path,rect,IMG_SIZE,VERBOSE)
        mask_image=process_image(meta.mask_path,rect,IMG_SIZE,VERBOSE)
        X.append(input_image)
        Y.append(mask_image)
        if cnt<10:
            plot_ioo(input_image,mask_image,None)
    except Exception as err:
        print("An exception occurred: {}".format(err))  
    #break

print ("Input images: {} Output images {}".format(len(X),len(Y)))
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size=0.7, shuffle=True)

def make_data_set (x_array,y_array):
    data_set = []
    for cnt,x in enumerate(x_array):
        data_set.append((x,y_array[cnt]))
    return data_set

train_dataset = make_data_set(x_train,y_train)
test_dataset = make_data_set(x_test,y_test)

print ("Test size: {}".format(len(test_dataset)))
print ("Train size: {}".format(len(train_dataset)))
```

```python
from torch import optim
from torch import nn
import torch
import torch.nn.functional as F
import torchvision

import numpy as np
import math
import gc

import torchvision.transforms as transforms
from torchvision.transforms.functional import to_grayscale

def execute (runnable):
    if torch.cuda.device_count()>0:
        print ("  --> Let's work with cuda")
        device = torch.device("cuda:0")
        gc.collect()
        torch.cuda.empty_cache()
        if runnable is not None:
            with torch.cuda.device(0):
                return runnable(device)
        return device
    else:
        print ("  --> Let's work with CPU")
        device = torch.device("cpu")
        if runnable is not None:
            return runnable(device)
        return device

for x in range (10):
    execute(None)
```

```python
class DatasetAdapter(torch.utils.data.Dataset):

    def __init__(self, dataset, from_cnt=0, to_cnt=1000):
        self.dataset = dataset
        self.from_cnt = from_cnt
        self.to_cnt = to_cnt

    def __len__(self):
        # TODO: Let's assert the negative values
        return self.to_cnt - self.from_cnt

    def __getitem__(self, idx):
        biased_index = self.from_cnt+idx
        image, label = self.dataset[biased_index]
        
        image = transforms.ToTensor()(image)
        image = image/image.max()

        label = transforms.ToTensor()(label)
        label = label/label.max()
        
        return image,label
```

```python
TRAIN_SIZE = 1000
train_set = DatasetAdapter(dataset=train_dataset,
                           from_cnt=0,to_cnt=TRAIN_SIZE)

TEST_SIZE = 500
test_set = DatasetAdapter(dataset=test_dataset,
                           from_cnt=0,to_cnt=TEST_SIZE)
```

```python
def to_device(data, device):
    if isinstance(data, (list,tuple)):
        return [to_device(x, device) for x in data]
    return data.to(device, non_blocking=True)

def load_to_device(loader,device):
    load_list=[]
    for i, data in enumerate(loader):
        load_list.append((to_device(data[0],device), to_device(data[1],device)))
    return load_list

BATCH_SIZE = int(TRAIN_SIZE/10)

train_set_loader = torch.utils.data.DataLoader(
        train_set,batch_size=BATCH_SIZE,shuffle=True)
test_set_loader = torch.utils.data.DataLoader(
        test_set, batch_size=TEST_SIZE,shuffle=True)

train_list = load_to_device(train_set_loader,execute(None))
test_list = load_to_device(test_set_loader,execute(None))
```

```python
def conv_block(in_channels,  out_channels, kernel_size =3):
    conv = nn.Sequential(
        nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, padding=(1,1), padding_mode='replicate'),
        nn.ReLU(),
        nn.BatchNorm2d(num_features=out_channels),
        # nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size),
        # nn.ReLU(),
        # nn.BatchNorm2d(num_features=out_channels)
    )
    return conv

class Unet(nn.Module):
    def __init__(self, num_classes):
        super(Unet, self).__init__()

        self.num_classes = num_classes
        self.down_conv_11 = conv_block(in_channels=CHANNEL_NUM, out_channels=IMG_SIZE)
        self.down_conv_12 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv_21 = conv_block(in_channels=IMG_SIZE, out_channels=IMG_SIZE*2)
        self.down_conv_22 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.down_conv_31 = conv_block(in_channels=IMG_SIZE*2, out_channels=IMG_SIZE*4)
        self.down_conv_32 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.middle = conv_block(in_channels=IMG_SIZE*4, out_channels=IMG_SIZE*8)

        self.up_conv_11 = nn.ConvTranspose2d(in_channels=IMG_SIZE*8, out_channels=IMG_SIZE*4,kernel_size=3, stride=2, padding=1, output_padding=1)
        #Disabling temporary
        #self.up_conv_12 = conv_block(in_channels=IMG_SIZE*8, out_channels=IMG_SIZE*4)
        self.up_conv_21 = nn.ConvTranspose2d(in_channels=IMG_SIZE*4, out_channels=IMG_SIZE*2, kernel_size=3, stride=2, padding=1, output_padding=1)
        #self.up_conv_22 = conv_block(in_channels=IMG_SIZE*4, out_channels=IMG_SIZE*2)
        self.up_conv_31 = nn.ConvTranspose2d(in_channels=IMG_SIZE*2, out_channels=IMG_SIZE*1, kernel_size=3, stride=2, padding=1, output_padding=1)
        #self.up_conv_32 = conv_block(in_channels=IMG_SIZE*2, out_channels=IMG_SIZE)
        self.output = nn.Conv2d(in_channels=IMG_SIZE, out_channels=num_classes, kernel_size=3, stride=1, padding=1)

    def forward(self, X):
        #print(X.shape)

        x11 = self.down_conv_11(X) 
        #print("l11",x11.shape)
        x12 = self.down_conv_12(x11)
        #print("l12",x12.shape)
        x21 = self.down_conv_21(x12)
        #print("l21",x21.shape)
        x22 = self.down_conv_22(x21)
        #print("l22",x22.shape)
        x31 = self.down_conv_31(x22)
        #print("l31",x31.shape)
        x32 = self.down_conv_32(x31)
        #print("l32",x32.shape)
        middle_out = self.middle(x32)

        # TODO: mixing with X was skipped
        y1 = self.up_conv_11(middle_out)
        y2 = self.up_conv_21(y1)
        y3 = self.up_conv_31(y2)
        Y = self.output(y3)
        return Y
```

```python
class DiceLoss(nn.Module):
    def __init__(self, weight=None, size_average=True):
        super(DiceLoss, self).__init__()

    def forward(self, logits, targets):
        smooth = 1
        num = targets.size(0)
        probs = logits
        m1 = probs.reshape(num, -1)
        m2 = targets.reshape(num, -1)
        intersection = (m1 * m2)

        score = (2. * intersection.sum(1) + smooth) / (m1.sum(1) + m2.sum(1) + smooth)
        score = 1 - (score.sum() / num)
        return score
```

```python
LIPS_CLASS_NUM =1

network = Unet(num_classes=LIPS_CLASS_NUM).to(execute(None))

for idx, module in enumerate(network.modules()):
     print(idx, '->', module)

#optimizer = torch.optim.Adam(network.parameters())
optimizer = optim.SGD(network.parameters(), lr=0.005, momentum=0.002)
criteria = DiceLoss()

```

```python
def adapt_perceptron (device):
    epoch_max_num = 1000
    max_test_tp=0
    history=[]
    
    for epoch_num in range(epoch_max_num):
        
        running_loss = 0.0
        for train_batch in train_list:
            x_train, y_train = train_batch[0], train_batch[1]
            optimizer.zero_grad()
            y_pred = network(x_train)
            loss = criteria(y_pred, y_train)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        print("Epoch [{}/{}] loss: {}".format(epoch_num,epoch_max_num,running_loss))
        history.append((epoch_num,running_loss))
            
    print('Adapting is finished!')
    return history
```

```python
train_history = execute(runnable=adapt_perceptron)
```

```python
epoch_list=[]
dice_list=[]
for (num,dice) in train_history:
    epoch_list.append(num)
    dice_list.append(dice)

plt.plot(epoch_list,dice_list)
plt.show()
```
