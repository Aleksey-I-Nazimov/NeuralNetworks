```python
import os

root_path = "../../../DataSetStore/set_lipstick_original"
#root_path = "/home/alex/dev/AiLearning/DataSetStore/set_lipstick_original"

class Meta:
    def __init__(self, image_file, mask_file, width, height,x_min,x_max,y_min,y_max):
        self.image_path = self.update_image_path(file_name=image_file)
        self.mask_path = self.update_mask_path(file_name=mask_file)
        self.width = int(width)
        self.height = int(height)
        self.x_mask_min = int(x_min)
        self.x_mask_max = int(x_max)
        self.y_mask_min = int(y_min)
        self.y_mask_max = int(y_max)

    def update_image_path (self,file_name):
        return os.path.join(root_path,"720p",file_name)

    def update_mask_path (self,file_name):
        return os.path.join(root_path,"mask",file_name)

    def calculate_suitable_size(self,lips_per_face):
        delta_x = self.x_mask_max-self.x_mask_min
        delta_y = self.y_mask_max-self.y_mask_min

        x1 = self.check_value(self.x_mask_min-delta_x)
        y1 = self.check_value(self.y_mask_min-int(lips_per_face*delta_y))
        x2 = self.check_value(self.x_mask_max+delta_x,self.width)
        y2 = self.check_value(self.y_mask_max+delta_y,self.height)
        return (y1,y2,x1,x2)

    def check_value(self,value,limit=None):
        if limit==None:
            if value<0:
                return 0
            else: return value
        else:
            if value>limit:
                return limit-1
            else:
                return value

    def __str__(self):
        max_str="coordinates: {} {} {} {}".format(self.x_mask_min,self.x_mask_max,self.y_mask_min,self.y_mask_max)
        return "Meta[image_path={},mask_path={},width={},height={},{}]".format(self.image_path,
                                                                            self.mask_path,
                                                                            self.width,
                                                                            self.height,
                                                                            max_str)


csv_path = os.path.join(root_path,"list.csv")
meta_list = []
with open(csv_path) as csv_file:
    file_lines = csv_file.read().splitlines()
    for cnt,line in enumerate(file_lines):
        if cnt==0:
            #print (line)
            continue
        #print (line)
        segments = line.split(",")
        meta=Meta(image_file=segments[0],
                  mask_file=segments[8],
                  width=segments[1],
                  height=segments[2],
                  x_min=segments[4],
                  x_max=segments[6],
                  y_min=segments[5],
                  y_max=segments[7])
        meta_list.append(meta)
        #print(meta)
        #break
```

```python
        
import cv2
import matplotlib.pyplot as plt
from pathlib import Path


def load_image_by_path(file_path,verbose=True):
    path = Path(file_path)
    if verbose :
        print ("  --> Loading {}".format(path.stat()))
        
    image= cv2.imread(str(path), cv2.IMREAD_UNCHANGED)
    
    if verbose :
        print (image.shape)
    return image

def to_gray_and_crop(image,rect,verbose=True):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    if verbose:
        print ("  --> Cropping: {}".format(rect))
        
    gray= gray[rect[0]:rect[1],rect[2]:rect[3]]
    return gray

def decrese_rect_size(image,rect_size,verbose=True):
    new_size = (rect_size,rect_size)
    if verbose:
        print ("  --> Downscaling: from {} to {}".format(image.shape,new_size))
    new_image = cv2.resize(image, new_size, interpolation = cv2.INTER_LINEAR)
    return new_image

def process_image (image_path,best_rect,best_size,verbose):
    image = load_image_by_path(image_path,verbose)
    image = to_gray_and_crop(image,best_rect,verbose)
    image = decrese_rect_size(image,best_size,verbose)
    return image

def plot_ioo (input_img,output_img_1,output_img_2):
    
    if output_img_2 is not None:
        fig,ax = plt.subplots(1,3)
    else:
        fig,ax = plt.subplots(1,2)

    ax[0].imshow(input_img)
    ax[0].set_title('input_img')
    ax[1].imshow(output_img_1)
    ax[1].set_title('output_img_1')
    
    if output_img_2 is not None:
        ax[2].imshow(output_img_2)
        ax[2].set_title('output_img_2')
    plt.show()


LIPS_PER_FACE = 4.5
IMG_SIZE = 64
CHANNEL_NUM = 1
VERBOSE = False

X = []
Y = []

for cnt,meta in enumerate(meta_list):
    if VERBOSE:
        print (meta)
    rect = meta.calculate_suitable_size(lips_per_face=LIPS_PER_FACE)
    try:
        input_image = process_image(meta.image_path,rect,IMG_SIZE,VERBOSE)
        mask_image=process_image(meta.mask_path,rect,IMG_SIZE,VERBOSE)
        X.append(input_image)
        Y.append(mask_image)
        if cnt<10:
            plot_ioo(input_image,mask_image,None)
    except Exception as err:
        print("An exception occurred: {}".format(err))  
    #break

print ("Input images: {} Output images {}".format(len(X),len(Y)))
```

```python
class DatasetAdapter(torch.utils.data.Dataset):

    def __init__(self, dataset, from_cnt=0, to_cnt=1000):
        self.dataset = dataset
        self.from_cnt = from_cnt
        self.to_cnt = to_cnt

    def __len__(self):
        # TODO: Let's assert the negative values
        return self.to_cnt - self.from_cnt

    def __getitem__(self, idx):
        biased_index = self.from_cnt+idx
        image, label = self.dataset[biased_index]
        
        image = transforms.ToTensor()(image)
        image = image/image.max()

        label = transforms.ToTensor()(label)
        label = label/label.max()
        
        return image,label
```

```python
# https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html



#  in_channels (int) – Number of channels in the input image

#  out_channels (int) – Number of channels produced by the convolution

#  kernel_size (int or tuple) – Size of the convolving kernel

#  stride (int or tuple, optional) – Stride of the convolution. Default: 1

#  padding (int, tuple or str, optional) – Padding added to all four sides of the input. Default: 0

#  padding_mode (str, optional) – 'zeros', 'reflect', 'replicate' or 'circular'. Default: 'zeros'

#  dilation (int or tuple, optional) – Spacing between kernel elements. Default: 1

#  groups (int, optional) – Number of blocked connections from input channels to output channels. Default: 1

#  bias (bool, optional) – If True, adds a learnable bias to the output. Default: True

#  groups controls the connections between inputs and outputs.
#  in_channels and out_channels must both be divisible by groups. For example,
#  At groups=1, all inputs are convolved to all outputs.
#  At groups=2, the operation becomes equivalent to having two conv layers side by side,
#  each seeing half the input channels and producing half the output channels, and both subsequently concatenated.
#  At groups= in_channels, each input channel is convolved with its own set of filters
#  (of size out_channelsin_channelsin_channelsout_channels​).

# ATTENTION:------------------------------------------------------------------------------------
# When groups == in_channels and out_channels == K * in_channels, where K is a positive integer,
#
# this operation is also known as a “depthwise convolution”.
#

def make_conv2d (in_channels,layer_size,stride,kernel_size,groups):
    
    print ("Made from: ch {} sz {} strd {} ksz {} grp {}".format(in_channels,layer_size,stride,kernel_size,groups))
    
    layer = nn.Conv2d(in_channels=in_channels,
                      out_channels=layer_size,
                      kernel_size=kernel_size,
                      stride=stride,
                      padding=(1,1),
                      padding_mode='replicate',
                      groups=groups)
    return layer

layer = make_conv2d(in_channels=3,layer_size=6,stride=1,kernel_size=3,groups=3)
layer.weight = torch.nn.Parameter(torch.ones_like(layer.weight))
layer.bias = torch.nn.Parameter(torch.zeros_like(layer.bias))

x1=torch.zeros(16)
x1[x1==0]=-2
x2=torch.zeros(16)
x2[x2==0]=0
x3=torch.ones(16)

x=torch.concat((x1,x2,x3)).reshape((3,4,4))
print (x)

y = layer(x)
print (y)
```

```python
# inspired by: https://www.kaggle.com/code/balraj98/unet-for-building-segmentation-pytorch

class Convolution2d(nn.Module):
    def __init__(self, in_channels, out_channels, groups, stride=1, kernel=3, drop_coef=0.10):
        super(Convolution2d, self).__init__()
        self.layer = make_conv2d(in_channels=in_channels, layer_size=out_channels,
                                 stride=stride, kernel_size=kernel, groups=groups)
        self.dropout = nn.Dropout(drop_coef)
        self.normalizer = nn.BatchNorm2d(out_channels)
        self.activation = nn.ReLU(inplace=True)

    def forward(self,x_batch):
        xtr = self.layer(x_batch)
        xtr = self.activation(xtr)
        xtr = self.dropout(xtr)
        xtr = self.normalizer(xtr)
        return xtr

class DepthwiseConvolution(nn.Module):
    def __init__(self, in_channels, out_channels, groups, kernel=3, drop_coef=0.10, is_double=False):
        super(DepthwiseConvolution, self).__init__()
        self.layer_1 = Convolution2d(in_channels,out_channels, groups, 1, kernel,drop_coef)
        self.layer_2 = None
        if is_double:
            self.layer_2 = Convolution2d(out_channels,out_channels, groups, 1, kernel,drop_coef)

    def forward(self, input_batch):
        layer1_batch = self.layer_1(input_batch)
        if self.layer_2 != None:
            layer2_batch = self.layer_2(layer1_batch)
            return layer2_batch
        return layer1_batch

class DownScaleConvolution(nn.Module):
    def __init__(self, in_channels, out_channels, groups, kernel=3, drop_coef=0.15):
        super(DownScaleConvolution, self).__init__()
        self.layer = Convolution2d(in_channels,out_channels,groups, 2, kernel, drop_coef)

    def forward(self, input_batch):
        layer_batch = self.layer(input_batch)
        return layer_batch

class UpscaleConvolution(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpscaleConvolution, self).__init__()
        self.layer = nn.ConvTranspose2d(in_channels,out_channels, kernel_size=2, stride=2)

    def forward(self, input_batch):
        layer_batch = self.layer(input_batch)
        return layer_batch
```

```python
class EncodeDecode(nn.Module):
    def __init__(self, in_channels, layer_size, layer_num, kernel=3,class_num=1,is_double_convolution=False):
        super(EncodeDecode, self).__init__()
        self.layers=[]
        
        # Down scaling:--------------------------------
        for cnt in range (layer_num):
            input_size = in_channels
            if cnt>0:
                input_size = layer_size
            # The group number is equal to the first input channel number!
            self.layers.append(DepthwiseConvolution(input_size,layer_size,in_channels,kernel,is_double=is_double_convolution))
            self.layers.append(DownScaleConvolution(layer_size,layer_size,in_channels,kernel))
        
        # # Up scaling:----------------------------------    
        for cnt in range (layer_num):
            # The group number is set to 1, beacause all color features were processed with in groups
            self.layers.append(UpscaleConvolution(layer_size,layer_size))
            self.layers.append(DepthwiseConvolution(layer_size,layer_size,1,kernel,is_double=is_double_convolution))
            
        # The final output layer:----------------------
        self.layers.append(make_conv2d(in_channels=layer_size,layer_size=class_num,stride=1, kernel_size=kernel, groups=1))
        self.layers.append(nn.Sigmoid())
            
        # Registering modules for torch:---------------
        for cnt,layer in enumerate(self.layers):
            key = "layer{}".format(cnt)
            self.add_module(key,layer)

    def forward(self, input_batch):
        for layer in self.layers:
            input_batch=layer(input_batch)
        return input_batch
```

```python
def adapt_perceptron (device):
    epoch_max_num = 80000
    max_test_tp=0
    history=[]
    
    for epoch_num in range(epoch_max_num):
        
        running_loss = 0.0
        for train_batch in train_list:
            x_train, y_train = train_batch[0], train_batch[1]
            optimizer.zero_grad()
            y_pred = network(x_train)
            #print (y_pred.shape,y_train.shape)
            loss = criteria(y_pred, y_train)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        if epoch_num % 100 == 0:
            print("Epoch [{}/{}] loss: {}".format(epoch_num,epoch_max_num,running_loss))
            
        history.append((epoch_num,running_loss))
            
    print('Adapting is finished!')
    return history
```

```python
def dice_coef_calculator(img1, img2):
    smooth = 1
    intersection = (img1 * img2)
    score = (2. * intersection.sum() + smooth) / (img1.sum() + img2.sum() + smooth)
    return score

cpu = torch.device("cpu")
cpu_network = network.to(cpu)
dice_list=[]
with torch.no_grad():
    for x_test,y_test in test_set:
        y_pred = cpu_network(x_test.reshape((1,x_test.shape[0],x_test.shape[1],x_test.shape[2])).to(cpu))
        plot_size = (IMG_SIZE,IMG_SIZE)
        tst = x_test.reshape(plot_size)
        valid = y_test.reshape(plot_size)
        prediction = y_pred.reshape(plot_size)
        dice_coef = dice_coef_calculator(valid,prediction)
        dice_list.append(dice_coef)
        if dice_coef>0.85:
            print ("Excellent:")
            plot_ioo(tst,valid,prediction)
        if dice_coef<0.15:
            print ("Not good:")
            plot_ioo(tst,valid,prediction)
            

plt.plot(dice_list)
plt.show()
```

```python
class UnetModel(nn.Module):
    def __init__(self, in_channels, layer_size, scale_num, kernel=3,class_num=1,is_double_conv=False):
        super(UnetModel, self).__init__()
        self.layers=[]
        
        # Down scaling:---------------------------------
        for cnt in range (scale_num):
            input_size = in_channels
            if cnt>0:
                input_size = layer_size
            # The group number is equal to the first input channel number!
            self.layers.append(DepthwiseConvolution(in_channels=input_size,out_channels=layer_size,
                                                    groups=in_channels,kernel=kernel,is_double=is_double_conv))
            self.layers.append(DownScaleConvolution(layer_size,layer_size,in_channels,kernel))

        self.layer_half_num = len(self.layers)
        
        # Up scaling:-----------------------------------    
        for cnt in range (scale_num):
            # The group number is set to 1, beacause all color features were processed with in groups
            self.layers.append(UpscaleConvolution(layer_size,layer_size))
            self.layers.append(DepthwiseConvolution(in_channels=layer_size+layer_size,out_channels=layer_size,
                                                    groups=1,kernel=kernel,is_double=is_double_conv))
            
        # The final output layer:-----------------------
        self.layers.append(make_conv2d(in_channels=layer_size,layer_size=class_num,stride=1, kernel_size=kernel, groups=1))
        self.layers.append(nn.Sigmoid())
            
        # Registering modules for torch:---------------
        for cnt,layer in enumerate(self.layers):
            key = "layer{}".format(cnt)
            self.add_module(key,layer)

    def forward(self, input_batch):
        down_scale_results = []
        for cnt,layer in enumerate(self.layers):
            if cnt<self.layer_half_num:
                input_batch = layer(input_batch)
                down_scale_results.insert(0,input_batch)
            else:
                if isinstance(layer,DepthwiseConvolution):
                    result_idx = cnt-self.layer_half_num
                    unet_input = torch.cat([input_batch, down_scale_results[result_idx]], dim=1)
                    input_batch=layer(unet_input)
                else :
                    input_batch=layer(input_batch)
        return input_batch
```

```python
def adapt_perceptron (device):
    epoch_max_num = 50000
    max_test_tp=0
    history=[]
    
    for epoch_num in range(epoch_max_num):
        
        running_loss = 0.0
        for train_batch in train_list:
            x_train, y_train = train_batch[0], train_batch[1]
            optimizer.zero_grad()
            y_pred = unetmodel(x_train)
            #print (y_pred.shape,y_train.shape)
            loss = criteria(y_pred, y_train)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        if epoch_num % 100 == 0:
            print("Epoch [{}/{}] error: {}".format(epoch_num,epoch_max_num,running_loss))
            
        history.append((epoch_num,running_loss))
            
    print('Adapting is finished!')
    return history
```
